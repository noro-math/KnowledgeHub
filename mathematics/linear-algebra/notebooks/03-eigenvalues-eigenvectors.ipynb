# Eigenvalues and Eigenvectors

This notebook explores eigenvalues and eigenvectors, fundamental concepts in linear algebra.

## Learning Objectives
- Understand eigenvalues and eigenvectors
- Compute eigenvalues and eigenvectors
- Visualize eigenvectors geometrically
- Explore applications in data analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA

plt.rcParams['figure.figsize'] = (12, 8)
np.random.seed(42)
```

## Definition

For a square matrix $A$, a non-zero vector $v$ is an **eigenvector** with **eigenvalue** $\lambda$ if:

$$Av = \lambda v$$

This means the transformation $A$ scales the vector $v$ by factor $\lambda$ without changing its direction.

```python
# Example: Find eigenvalues and eigenvectors
A = np.array([[3, 1],
              [0, 2]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Matrix A:\n{A}")
print(f"\nEigenvalues: {eigenvalues}")
print(f"\nEigenvectors:\n{eigenvectors}")

# Verify the eigenvalue equation
for i in range(len(eigenvalues)):
    λ = eigenvalues[i]
    v = eigenvectors[:, i]
    
    Av = A @ v
    λv = λ * v
    
    print(f"\nFor eigenvalue {λ:.3f}:")
    print(f"Av = {Av}")
    print(f"λv = {λv}")
    print(f"Equal? {np.allclose(Av, λv)}")
```

## Geometric Visualization

```python
# Visualize eigenvectors and their transformation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Original eigenvectors
colors = ['red', 'blue']
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    ax1.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
               color=colors[i], width=0.005, label=f'v{i+1}')
    
    # Show transformed eigenvectors
    Av = A @ v
    ax2.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1, 
               color=colors[i], width=0.005, label=f'Av{i+1} = {eigenvalues[i]:.1f}v{i+1}')

# Add some other vectors to show general transformation
test_vectors = np.array([[1, 0], [0, 1], [1, 1], [-1, 1]]).T
transformed_vectors = A @ test_vectors

for i in range(test_vectors.shape[1]):
    v_orig = test_vectors[:, i]
    v_trans = transformed_vectors[:, i]
    
    ax1.quiver(0, 0, v_orig[0], v_orig[1], angles='xy', scale_units='xy', scale=1, 
               color='gray', alpha=0.5, width=0.003)
    ax2.quiver(0, 0, v_trans[0], v_trans[1], angles='xy', scale_units='xy', scale=1, 
               color='gray', alpha=0.5, width=0.003)

# Setup plots
for ax, title in zip([ax1, ax2], ['Original Vectors', 'Transformed Vectors']):
    ax.set_xlim(-2, 4)
    ax.set_ylim(-2, 3)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.set_title(title)
    ax.legend()

plt.tight_layout()
plt.show()
```

## Principal Component Analysis (PCA)

PCA uses eigenvalues and eigenvectors to find the principal directions of variation in data.

```python
# Generate sample data
X, _ = make_blobs(n_samples=200, centers=1, n_features=2, 
                  random_state=42, cluster_std=2.0)

# Add correlation by transforming the data
transform_matrix = np.array([[1, 0.8], [0, 1]])
X = X @ transform_matrix.T

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Get principal components (eigenvectors of covariance matrix)
components = pca.components_
explained_variance = pca.explained_variance_

print(f"Explained variance: {explained_variance}")
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Original data with principal components
ax1.scatter(X[:, 0], X[:, 1], alpha=0.6, s=30)
mean_point = np.mean(X, axis=0)

# Plot principal components
for i, (component, variance) in enumerate(zip(components, explained_variance)):
    # Scale by square root of eigenvalue for visualization
    scaled_component = component * np.sqrt(variance) * 2
    ax1.arrow(mean_point[0], mean_point[1], 
              scaled_component[0], scaled_component[1],
              head_width=0.3, head_length=0.3, fc=f'C{i}', ec=f'C{i}',
              label=f'PC{i+1} (var={variance:.2f})')

ax1.set_title('Original Data with Principal Components')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_aspect('equal')

# Transformed data (in PC space)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.6, s=30)
ax2.set_title('Data in Principal Component Space')
ax2.set_xlabel('First Principal Component')
ax2.set_ylabel('Second Principal Component')
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```

## Characteristic Polynomial

Eigenvalues are roots of the characteristic polynomial: $\det(A - \lambda I) = 0$

```python
# Example: Compute characteristic polynomial manually
A = np.array([[2, 1],
              [1, 2]])

print(f"Matrix A:\n{A}")

# For 2x2 matrix [[a,b],[c,d]], characteristic polynomial is:
# λ² - (a+d)λ + (ad-bc) = 0
a, b = A[0, 0], A[0, 1]
c, d = A[1, 0], A[1, 1]

trace = a + d
determinant = a*d - b*c

print(f"\nTrace (a+d): {trace}")
print(f"Determinant (ad-bc): {determinant}")
print(f"Characteristic polynomial: λ² - {trace}λ + {determinant} = 0")

# Solve quadratic equation
discriminant = trace**2 - 4*determinant
λ1 = (trace + np.sqrt(discriminant)) / 2
λ2 = (trace - np.sqrt(discriminant)) / 2

print(f"\nEigenvalues from formula: {λ1:.3f}, {λ2:.3f}")

# Compare with numpy
eigenvals_numpy = np.linalg.eigvals(A)
print(f"Eigenvalues from numpy: {eigenvals_numpy}")
```
