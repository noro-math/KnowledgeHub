# Vector Spaces

This notebook explores the fundamental concepts of vector spaces in linear algebra.

## Learning Objectives

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set up plotting style
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)
```

## Definition of Vector Space

A **vector space** $V$ over a field $F$ is a set equipped with two operations:
- Vector addition: $+: V \times V \to V$
- Scalar multiplication: $\cdot: F \times V \to V$

These operations must satisfy the following axioms...

```python
# Example: R^2 as a vector space
v1 = np.array([2, 3])
v2 = np.array([1, -1])
scalar = 2.5

# Vector addition
v_sum = v1 + v2
print(f"v1 + v2 = {v_sum}")

# Scalar multiplication
v_scaled = scalar * v1
print(f"{scalar} * v1 = {v_scaled}")
```

## Visualization of Vector Operations

```python
# Visualize vectors in R^2
fig, ax = plt.subplots(1, 1, figsize=(8, 6))

# Plot vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
ax.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='green', label='v1 + v2')

# Set up the plot
ax.set_xlim(-1, 4)
ax.set_ylim(-2, 4)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.legend()
ax.set_title('Vector Addition in R²')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
