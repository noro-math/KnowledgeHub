```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import matplotlib.animation as animation

plt.rcParams['figure.figsize'] = (12, 8)
```

## Definition

A function $T: V \to W$ between vector spaces is called a **linear transformation** if:

1. $T(u + v) = T(u) + T(v)$ for all $u, v \in V$
2. $T(cv) = cT(v)$ for all $c \in \mathbb{F}$ and $v \in V$

These can be combined into: $T(au + bv) = aT(u) + bT(v)$

```python
# Example: Rotation transformation in R^2
def rotation_matrix(theta):
    """Create a 2D rotation matrix for angle theta (in radians)"""
    return np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])

# 45-degree rotation
theta = np.pi / 4
R = rotation_matrix(theta)
print(f"Rotation matrix (45°):\n{R}")

# Apply to some vectors
vectors = np.array([[1, 0], [0, 1], [1, 1], [-1, 1]]).T
rotated_vectors = R @ vectors

print(f"\nOriginal vectors:\n{vectors}")
print(f"\nRotated vectors:\n{rotated_vectors}")
```

## Visualization of Linear Transformations

```python
# Visualize the effect of linear transformation on a unit square
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Original unit square
square = np.array([[0, 1, 1, 0, 0],
                   [0, 0, 1, 1, 0]])

# Apply transformation
A = np.array([[2, 1],
              [0, 1]])
transformed_square = A @ square

# Plot original
ax1.plot(square[0], square[1], 'b-', linewidth=2, label='Unit Square')
ax1.fill(square[0], square[1], alpha=0.3, color='blue')
ax1.grid(True, alpha=0.3)
ax1.set_aspect('equal')
ax1.set_title('Original')
ax1.set_xlim(-0.5, 2.5)
ax1.set_ylim(-0.5, 2.5)

# Plot transformed
ax2.plot(transformed_square[0], transformed_square[1], 'r-', linewidth=2, label='Transformed')
ax2.fill(transformed_square[0], transformed_square[1], alpha=0.3, color='red')
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')
ax2.set_title('After Transformation')
ax2.set_xlim(-0.5, 3.5)
ax2.set_ylim(-0.5, 2.5)

plt.tight_layout()
plt.show()

print(f"Transformation matrix:\n{A}")
print(f"Determinant (area scaling factor): {np.linalg.det(A)}")
```

## Kernel and Image

For a linear transformation $T: V \to W$:
- **Kernel (Null Space)**: $\ker(T) = \{v \in V : T(v) = 0\}$
- **Image (Range)**: $\text{im}(T) = \{T(v) : v \in V\}$

```python
# Example: Find kernel and image of a transformation
A = np.array([[1, 2, 3],
              [2, 4, 6],
              [1, 1, 2]])

print(f"Matrix A:\n{A}")
print(f"\nRank of A: {np.linalg.matrix_rank(A)}")

# Find null space (kernel)
U, s, Vt = np.linalg.svd(A)
null_space = Vt[np.abs(s) < 1e-10]
print(f"\nNull space basis:\n{null_space}")

# Verify it's in the kernel
if len(null_space) > 0:
    test_vector = null_space[0]
    result = A @ test_vector
    print(f"\nA * null_vector = {result}")
    print(f"Is approximately zero: {np.allclose(result, 0)}")
```
